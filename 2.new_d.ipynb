```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'text':['pepole watch campusx','campusx watch campusx','pepole write comment','campusx write comment'],'output':[1,1,0,0]})
df
```

```python
pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])   ### crfeat a object
```

```python
### vocabulary (find unique words)
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
```

### Out of vocabulray problem is solved using this method (BOW)

```python
cv.transform(['campusx watch and write  comment of campusx']).toarray()  ## here more words to be added
```

### N GRAMs

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
nbow = cv.fit_transform(df['text'])   ### crfeat a object
```

```python
print(cv.vocabulary_)   ## here we are getting 2 words bigrams
```
